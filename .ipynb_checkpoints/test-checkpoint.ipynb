```python
from __future__ import division
import osmgraph
import geog
from astar import A_star
from xml_creation import nodes_edges_to_xml
import geojsonio
import json
```

```python
filename = '/Users/grigoriipogorelov/Desktop/map_inno.osm'
g = osmgraph.parse_file(filename)

valid_nodes = {}
for n1, n2 in g.edges():
    valid_nodes[n1] = 0
    valid_nodes[n2] = 0
    c1, c2 = osmgraph.tools.coordinates(g, (n1, n2))
    g[n1][n2]['length'] = geog.distance(c1, c2)
```

```python
a_star = A_star(g, valid_nodes)
```

```python
srcs = [4895039589]
dsts = [3621314698]
k_range = [1,2]
for src in srcs:
    for dst in dsts:
        a_star_path, _ = a_star.a_star_search(src,dst, k_range)
```

```python
nodes_edges_to_xml(g, a_star_path)
```

```python
coords = osmgraph.tools.coordinates(g, a_star_path)

geojsonio.display(json.dumps([{'type': 'LineString', 'coordinates': coords}]))
```
